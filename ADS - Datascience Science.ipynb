```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setting the style for clearer visualization
sns.set(style="whitegrid")

# Load the dataset
file_path = "/content/DataScience_salaries_2024.csv"
data = pd.read_csv(file_path)

# Convert salary column to numeric in case of any issues with data type
data['salary_in_usd'] = pd.to_numeric(data['salary_in_usd'], errors='coerce')

# 1. Bar Chart: Average Salary by Experience Level and Employment Type


def plot_avg_salary_by_experience_and_employment_type():
    """
    Plots a bar chart showing the average salary in USD by experience level and employment type.
    """
    avg_salary_exp_type = data.groupby(['experience_level', 'employment_type'])[
        'salary_in_usd'].mean().reset_index()
    avg_salary_exp_type = avg_salary_exp_type.sort_values(
        by='salary_in_usd', ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x='salary_in_usd', y='experience_level',
                hue='employment_type', data=avg_salary_exp_type, palette='coolwarm')
    plt.title(
        'Average Salary by Experience Level and Employment Type (in USD)', fontsize=14)
    plt.xlabel('Average Salary (USD)', fontsize=12)
    plt.ylabel('Experience Level', fontsize=12)
    plt.legend(title='Employment Type',
               bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# 2. Box Plot: Salary Distribution by Experience Level


def plot_salary_by_experience_level():
    """
    Plots a box plot showing the distribution of salaries in USD for each experience level.
    """
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='experience_level', y='salary_in_usd',
                data=data, palette='Set2')
    plt.title('Salary Distribution by Experience Level', fontsize=14)
    plt.xlabel('Experience Level', fontsize=12)
    plt.ylabel('Salary (USD)', fontsize=12)
    plt.show()

# 3. Line Plot: Salary Trend Over Years


def plot_salary_trend_over_years():
    """
    Plots a line chart showing the average salary in USD per year.
    """
    yearly_avg_salary = data.groupby(
        'work_year')['salary_in_usd'].mean().reset_index()

    plt.figure(figsize=(10, 6))
    sns.lineplot(x='work_year', y='salary_in_usd',
                 data=yearly_avg_salary, marker='o', color='blue')
    plt.title('Average Salary Trend Over Years (in USD)', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Average Salary (USD)', fontsize=12)
    plt.xticks(yearly_avg_salary['work_year'])
    plt.show()


# Plot the graphs
plot_avg_salary_by_experience_and_employment_type()
plot_salary_by_experience_level()
plot_salary_trend_over_years()
```
